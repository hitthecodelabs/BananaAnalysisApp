```python
import cv2, os
import numpy as np
from glob import glob

from PIL import Image, ImageDraw, ImageFilter, ImageOps
```

```python
from scipy import ndimage
from skimage import measure
```

```python
exportacion = glob("../ex*/*")
rechazo = glob("../re*/*")
```

```python
def crop_banana(image_path):
    
    # Leer la imagen
    img = cv2.imread(image_path)

    # Convertir a escala de grises
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Aplicar binarización
    _, binary = cv2.threshold(gray, 
                              150, ### balance de blancos
                              255, ### iluminacion
                              cv2.THRESH_BINARY_INV)

    # Encontrar contornos en la imagen binarizada
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Ordenar los contornos por área (asumiendo que la mano de banano es el contorno más grande)
    # al momento de ordenar, se ordena por tamaño de contorno
    # desde el contorno mas grande al pequeño
    contours = sorted(contours, key=cv2.contourArea, reverse=True)

    # Obtener el rectángulo delimitador del contorno más grande
    x, y, w, h = cv2.boundingRect(contours[0]) ### 0 indica el primer contorno, es decir, el mas grande

    # Recortar la imagen utilizando las coordenadas del rectángulo delimitador
    cropped = img[y:y+h, x:x+w]
    cropped_rgb = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)

    return cropped_rgb

def crop_banana2(image_path):
    # Read the image
    img = Image.open(image_path)
    img_np = np.array(img)

    # Convert image to grayscale
    img_gray = ImageOps.grayscale(img)

    # Apply threshold
    img_binary = img_gray.point(lambda p: 255 if p < 150 else 0)

    # Convert to NumPy array to find contours
    img_binary_np = np.array(img_binary)

    # Find contours (using OpenCV, since PIL doesn't have this function)
    contours, _ = cv2.findContours(img_binary_np, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Sort contours by area
    contours = sorted(contours, key=cv2.contourArea, reverse=True)

    # Get the bounding box of the largest contour
    x, y, w, h = cv2.boundingRect(contours[0])

    # Crop the image using bounding box coordinates
    cropped = img_np[y:y+h, x:x+w]

    return cropped

def crop_banana3(image_path):
    # Read the image
    img = Image.open(image_path)
    img_np = np.array(img)

    # Convert image to grayscale
    img_gray = ImageOps.grayscale(img)

    # Apply threshold
    img_binary = img_gray.point(lambda p: 255 if p < 150 else 0)

    # Convert to NumPy array to find contours
    img_binary_np = np.array(img_binary)

    # Label the binary image
    labeled_array, num_features = ndimage.label(img_binary_np)

    # Find contours using skimage
    contours = measure.find_contours(labeled_array, 0.8)

    # Sort contours by area (size)
    contours = sorted(contours, key=lambda x: len(x), reverse=True)

    # Get bounding box of the largest contour
    x_min = int(np.min(contours[0][:, 1]))
    x_max = int(np.max(contours[0][:, 1]))
    y_min = int(np.min(contours[0][:, 0]))
    y_max = int(np.max(contours[0][:, 0]))

    # Crop the image using bounding box coordinates
    cropped = img_np[y_min:y_max, x_min:x_max]

    return cropped

def redim(image, alto=272, largo=416): ### redimensionamiento de la imagen
    
    # Obtener las dimensiones de la imagen recortada
    old_height, old_width = image.shape[:2]

    # Calcular la relación de aspecto de la nueva imagen y la antigua
    aspect_ratio = old_width / old_height

    # Rellenar la imagen si es necesario para mantener la relación de aspecto
    if old_width < old_height * (largo / alto):
        new_width = int(old_height * (largo / alto))
        padding = (new_width - old_width) // 2
        padded = cv2.copyMakeBorder(image, 0, 0, padding, padding, cv2.BORDER_CONSTANT, value=[0, 0, 0])
    else:
        new_height = int(old_width / (largo / alto))
        padding = (new_height - old_height) // 2
        padded = cv2.copyMakeBorder(image, padding, padding, 0, 0, cv2.BORDER_CONSTANT, value=[0, 0, 0])

    # Redimensionar la imagen al tamaño deseado
    resized = cv2.resize(padded, (largo, alto))

    return resized

def redim2(image, alto=272, largo=416):
    # Convert the NumPy array to a PIL image first
    image = Image.fromarray(image)

    # Get the dimensions of the cropped image
    old_width, old_height = image.size

    # Calculate the aspect ratio of the old and new image
    aspect_ratio = old_width / old_height

    # Pad the image to maintain aspect ratio
    if old_width < old_height * (largo / alto):
        new_width = int(old_height * (largo / alto))
        padding = (new_width - old_width) // 2
        padded = ImageOps.expand(image, (padding, 0, padding, 0), fill="black")
    else:
        new_height = int(old_width / (largo / alto))
        padding = (new_height - old_height) // 2
        padded = ImageOps.expand(image, (0, padding, 0, padding), fill="black")

    # Resize the image to the desired size
    resized = padded.resize((largo, alto), Image.LANCZOS)

    # Convert the PIL image back to a NumPy array if needed
    resized_np = np.array(resized)

    return resized_np
```
